```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Importing of data from excel sheets
AprilSales = pd.read_csv("dataset/201904 sales reciepts.csv")
customer = pd.read_csv("dataset/customer.csv")
product = pd.read_csv("dataset/product.csv")
generations = pd.read_csv("dataset/generations.csv")
customer.head()
```

```python
# Merging of sales and customer data sets
merge1 = AprilSales.merge(customer, how='left')
print(AprilSales.shape)
print(customer.shape)
print(merge1.shape)
merge1.head()
```

```python
# Merging of sales, customer, and product data sets
merge2 = merge1.merge(product, how='inner')
print(merge1.shape)
print(merge2.shape)
merge2.head()
```

```python
# Merging of sales, customer, product, and generations data sets
merge3 = merge2.merge(generations, how='left')
print(merge2.shape)
print(merge3.shape)
merge3.head()
```

```python
# What kinds of products generate the most sales?

# Creating a donut chart for product group sales
labels = ['Add-ons', 'Beverages', 'Food','Merchandise','Whole Bean/Teas']
colors = ['red', 'blue', 'green', 'yellow','orange']
percentage = grouped3['line_item_amount']/grouped3['line_item_amount'].sum()*100
percentage_r = round(percentage,1)
fig, ax = plt.subplots()
wedges, texts = ax.pie(grouped3['line_item_amount'], startangle=0,colors=colors)
my_circle = plt.Circle((0, 0), 0.7, color='white')
p = plt.gcf()
p.gca().add_artist(my_circle)

# Calculate angles and positions for annotations
for i, wedge in enumerate(wedges):
    angle = (wedge.theta1 + wedge.theta2) / 2
    x = np.cos(np.deg2rad(angle))
    y = np.sin(np.deg2rad(angle))
    
    # Adjust label position
    label_x = 1.3 * x
    label_y = 1.3 * y

    # Create annotation
    ax.annotate(
        f'{percentage_r[i]}%',
        xy=(x, y),
        xytext=(label_x, label_y),
        arrowprops=dict(arrowstyle='-', color='black'),
        ha='center',
        va='center'
    )
plt.title("Total Sales Breakdown by Product Group", fontsize=14)
plt.legend(labels, loc='upper right', bbox_to_anchor=(1.4, 0.95))
plt.savefig('Fig1.jpg', bbox_inches='tight',dpi=300)
plt.show()
```

```python
# Calculating RFM Metrics
import datetime as dt
AprilSales["transaction_date"] = pd.to_datetime(AprilSales["transaction_date"])
AprilSales["transaction_date"].max()
today_date = dt.datetime(2019, 5, 1)
rfm = AprilSales.groupby("customer_id").agg({"transaction_date": lambda transaction_date: (today_date - transaction_date.max()).days,
                                     "transaction_id": lambda transaction_id: transaction_id.nunique(),
                                     "line_item_amount": lambda line_item_amount: line_item_amount.sum()})
rfm.columns = ["Recency", "Frequency", "Monetary"]
rfm.describe().T
```

```python
# Score each column (1–5)
rfm["R_score"] = pd.qcut(rfm["Recency"].rank(method="first"), 5, labels=[5, 4, 3, 2, 1])
rfm["F_score"] = pd.qcut(rfm["Frequency"], 5, labels=[1, 2, 3, 4, 5])
rfm["M_score"] = pd.qcut(rfm["Monetary"], 5, labels=[1, 2, 3, 4, 5])
rfm["RF_Score"] = rfm["R_score"].astype(str) + rfm["F_score"].astype(str)
seg_map = {
    r"[1-2][1-2]": "Hibernating",
    r"[1-2][3-4]": "At Risk",
    r"[1-2]5": "Can't Lose Them",
    r"3[1-2]": "About to Sleep",
    r"33": "Need Attention",
    r"[3-4][4-5]": "Loyal Customers",
    r"41": "Promising",
    r"51": "New Customers",
    r"[4-5][2-3]": "Potential Loyalists",
    r"5[4-5]": "Champions"
}
rfm["segment"] = rfm["RF_Score"].replace(seg_map, regex=True)
rfm=rfm.reset_index()
rfm.head()
```

```python
# Merging of product data with RFM data
rfmgen = dataclean4.merge(rfm, how='inner')
print(dataclean4.shape)
print(rfmgen.shape)
rfmgen.head()
```
